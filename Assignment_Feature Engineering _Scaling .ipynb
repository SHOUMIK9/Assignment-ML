```python
import pandas as pd
import numpy as np
```

```python
data= {'Price' : [110,105,115,120,110,130,150,100,105 ]}
df= pd.DataFrame(data)
```

```python
df
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
df['Price_Normalization']=scaler.fit_transform(df[['Price']])
```

```python
df
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
stand = StandardScaler()
```

```python
df['Price_Standardization'] = stand.fit_transform(df[['Price']])
```

```python
df
```

# Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
df['Price_Log'] = FunctionTransformer(np.log1p).fit_transform(df[['Price']])
```

```python
df
```

# Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
robust = RobustScaler()
```

```python
df['Price_Robust'] = robust.fit_transform(df[['Price']])
```

```python
df
```

# Max Absolute Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
maxabs = MaxAbsScaler()
```

```python
df['Price_MaxAbs'] = maxabs.fit_transform(df[['Price']]) 
```

```python
df
```
